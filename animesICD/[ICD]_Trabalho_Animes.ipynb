## Contagem de Animes por Temas (My Anime List)

Quais os temas mais comuns abordados por animes?

```python
def criar_dataframe_temas_my_anime_list():
  df_filtrado = df_animes_anime_list[["Title", "English", "Themes", "Type", "Demographics", "Start_Aired", "Favorites", "Members", "Score", "Scored_Users"]].copy()
  df_filtrado = df_filtrado[df_filtrado["Themes"] != "Unknown"].copy()
  df_filtrado = df_filtrado[df_filtrado["Type"] != "Music"].copy()

  df_filtrado["Themes"] = df_filtrado["Themes"].str.split(",")
  df_filtrado = df_filtrado.explode("Themes", ignore_index = True)
  df_filtrado["Themes"] = df_filtrado["Themes"].str.strip()

  return df_filtrado.copy()
```

```python
df_temas = criar_dataframe_temas_my_anime_list()
df_temas.info()
```

```python
def plot_contagem_animes_por_tema():
  df_contagem_animes_por_tema = (
                                  df_temas[["Themes", "Title"]].
                                    groupby("Themes").
                                    count().
                                    rename(columns = {"Title": "Quantidade_Animes"}).
                                    sort_values(by="Quantidade_Animes", ascending = False).
                                    reset_index()
                                )

  plot = sns.barplot(df_contagem_animes_por_tema, x="Quantidade_Animes", y="Themes", color="steelblue")

  plot.set(title="Quantidade de Animes que abordam determinado Tema")
  plot.set_ylabel("Tema")
  plot.set_xlabel("Quantidade de Animes")

  plt.show()
```

```python
plot_contagem_animes_por_tema()
```

## Contagem de Animes por Gênero (My Anime List)

Quais os gêneros mais comuns em animes?

```python
def criar_dataframe_genero_my_anime_list():
  df_filtrado = df_animes_anime_list[["Title", "English", "Genres", "Type", "Demographics", "Favorites", "Members", "Score", "Scored_Users"]].copy()
  df_filtrado = df_filtrado[df_filtrado["Genres"] != "Unknown"].copy()
  df_filtrado = df_filtrado[df_filtrado["Type"] != "Music"].copy()

  df_filtrado["Genres"] = df_filtrado["Genres"].str.split(",")
  df_filtrado = df_filtrado.explode("Genres", ignore_index = True)
  df_filtrado["Genres"] = df_filtrado["Genres"].str.strip()

  return df_filtrado.copy()
```

```python
def plot_contagem_animes_por_genero():
  df_generos = criar_dataframe_genero_my_anime_list()

  df_contagem_animes_por_genero = (
                                    df_generos[["Genres", "Title"]].
                                      groupby("Genres").
                                      count().
                                      rename(columns = {"Title": "Quantidade_Animes"}).
                                      sort_values(by="Quantidade_Animes", ascending = False).
                                      reset_index()
                                  )

  plot = sns.barplot(df_contagem_animes_por_genero, x="Quantidade_Animes", y="Genres", color="steelblue")

  plot.set(title="Quantidade de Animes de determinado Gênero")
  plot.set_ylabel("Tema")
  plot.set_xlabel("Quantidade de Animes")

  plt.show()
```

## Genêros mais bem avaliados

Podemos comparar os animes por meio de 3 métricas de avaliação:

*   **Score**: Nota atribuida com base nas avaliações dos usuários (Scored_Users é a quantidade de usuários que deram uma nota para determinado anime)

*   **Members**: A quantidade de usuários que adicionaram aquele anime a sua lista do My Anime Liste

*   **Favorites**: A quantidade de usuários que marcaram aquele anime como favorito

```python
def descrever_animes_por_genero_e(coluna_alvo):
  df_generos = criar_dataframe_genero_my_anime_list()
  return df_generos[["Genres", coluna_alvo]].groupby("Genres").describe()
```

### Avaliando Gêneros por Média

```python
def plot_bar_por_genero(coluna, operacao):
  df_generos = criar_dataframe_genero_my_anime_list()
  df_medias_generos = (
                        df_generos.
                          groupby("Genres").
                          agg(operacao, numeric_only = True).
                          reset_index()
                      )

  df_medias_scores_generos = df_medias_generos[["Genres", coluna]].sort_values(by = coluna, ascending = False)
  plot = sns.barplot(df_medias_scores_generos, x="Genres", y=coluna, color="steelblue")

  plt.xticks(rotation=45)
  plt.show()
```

## Avaliando Gêneros por Demografias

Uma questão muito comum em animes e produções japonesas é separar as obras por demografias. Elas são agrupadas da seguinte forma:

*   **Shounen:** São as produções voltadas ao público masculino jovem/adoslecente
*   **Shoujo:** São as produções voltadas ao público feminino jovem/adoslecente

*   **Seinen:** São as produções voltadas ao público masculino adulto
*   **Josei:** São as produções voltadas ao público feminino adulto

*   **Kids:** São as produções para o público infantil

Nós comparamos os Scores médios de cada Gênenro para cada Demografia. Além disso, realizamos o plot das demografias voltadas a uma faixa étaria em comum.

```python
def criar_dataframe_separado_genero_demografia_my_anime_list():
  df_filtrado = df_animes_anime_list[["Title", "English", "Genres", "Type", "Demographics", "Favorites", "Members", "Score", "Scored_Users"]].copy()
  df_filtrado = df_filtrado[df_filtrado["Genres"] != "Unknown"].copy()
  df_filtrado = df_filtrado[df_filtrado["Type"] != "Music"].copy()

  # Expande linhas por gênero
  df_filtrado["Genres"] = df_filtrado["Genres"].str.split(",")
  df_filtrado = df_filtrado.explode("Genres", ignore_index = True)
  df_filtrado["Genres"] = df_filtrado["Genres"].str.strip()

  # Expande linhas por demografia
  df_filtrado["Demographics"] = df_filtrado["Demographics"].str.split(",")
  df_filtrado = df_filtrado.explode("Demographics", ignore_index = True)
  df_filtrado["Demographics"] = df_filtrado["Demographics"].str.strip()

  return df_filtrado.copy()
```

### Realiza comparação do Score de Gênero para as Demografias jovens


```python
def plot_line_baseado_grupo_demografia(demografia_1, demografia_2):
  df_separado_em_genero_demografia = criar_dataframe_separado_genero_demografia_my_anime_list()

  filtro_demografia_1 = (df_separado_em_genero_demografia["Demographics"] == demografia_1)
  filtro_demografia_2 = (df_separado_em_genero_demografia["Demographics"] == demografia_2)

  df_demografia = df_separado_em_genero_demografia[filtro_demografia_1 | filtro_demografia_2]

  dados_plot_demografia = (
                            df_demografia.
                              groupby(["Demographics", "Genres"]).
                              mean(numeric_only = True)
                          )

  plot = sns.lineplot(data=dados_plot_demografia, x="Genres", y="Score", hue="Demographics", marker="o")
  plt.xticks(rotation=45)
  plt.show()

  dados_plot_demografia = dados_plot_demografia.reset_index()

  plot = sns.barplot(data=dados_plot_demografia, x="Genres", y="Score", hue="Demographics")
  plt.xticks(rotation=45)
  plt.show()

  return
```

```python
plot_line_baseado_grupo_demografia("Shounen", "Shoujo")
```

# Intervalo Confiança ou Teste de Permutação

1) Intervalo de confiança para a média do tempo de duração de episódio dos animes ou quantidade de episódios animes  (Tratar para o tempo de duração os animes que são filmes Type == Movies)
ou
2) Teste de permutação para verificar se: Vemos que suspense tem mais favoritos/membros em média. O valor presente nele, ao fazer o testes de permutação separando entre suspense e não suspense (ou agrupar por gêneros semelhantes como Drama e Mystery), está dentro da normal/faixa indicaria que isso poderia ser explicado pelo acaso e suspense não é tão grande coisa assim.
ou
3) Outra que você achar que rola

```python
#IC com 95% de significancia para o tempo médio de duraçao dos episódios de cada tipo
import math

tipos = ['TV', 'Movie', 'ONA', 'OVA', 'Special', 'Music']
for tipo in tipos :

  df_tipo = df_animes_anime_list[df_animes_anime_list['Type'] == tipo]
  tempo_medio = np.zeros(5000)

  for i in range(5000) :
    df_tipo_novo = df_tipo.sample(n=len(df_tipo), replace=True)
    tempo_medio[i] = df_tipo_novo['Duration_Minutes'].mean()

  LI = np.percentile(tempo_medio, 2.5)
  LS = np.percentile(tempo_medio, 97.5)

  if (LI - math.floor(LI) >= 0.60) :
    LI = math.floor(LI) + 1 + (LI - math.floor(LI) - 0.6)
  if (LS - math.floor(LS) >= 0.60) :
    LS = math.floor(LS) + 1 + (LS - math.floor(LS) - 0.6)

  print(tipo, "\n", "IC: ",(LI.round(2),LS.round(2)))
```

```python
def ajustar_df_animes_suspense():
  def identificar_animes_suspense(linha):
    genero_anime = linha["Genres"]

    if(genero_anime == "Suspense"):
      return 1

    return 0

  df_genero_animes = criar_dataframe_genero_my_anime_list()
  df_filtrado = df_genero_animes[["Genres", "Favorites"]].copy()
  df_filtrado["identificador_suspense"] = df_filtrado.apply(identificar_animes_suspense, axis=1)

  return df_filtrado.copy()
```

```python
def get_estatistica_observada_animes_suspense():
  df_filtrado = ajustar_df_animes_suspense()
  df_medias = df_filtrado.groupby("identificador_suspense").mean(numeric_only = True).reset_index()

  media_animes_suspense = df_medias.loc[df_medias["identificador_suspense"] == 1, "Favorites"].iloc[0]
  media_animes_nao_suspense = df_medias.loc[df_medias["identificador_suspense"] == 0, "Favorites"].iloc[0]

  return (media_animes_suspense - media_animes_nao_suspense)
```

```python
get_estatistica_observada_animes_suspense()
```

```python
def aplicar_teste_permutacao_animes_suspense():
  QUANTIDADE_TESTES = 10000

  df_filtrado = ajustar_df_animes_suspense()
  coluna_flag_suspense = df_filtrado["identificador_suspense"].values

  medias_diferenca = np.zeros(QUANTIDADE_TESTES)

  for i in range(QUANTIDADE_TESTES):
    np.random.shuffle(coluna_flag_suspense)

    df_filtrado["identificador_suspense"] = coluna_flag_suspense
    df_medias = df_filtrado.groupby("identificador_suspense").mean(numeric_only = True).reset_index()

    media_casas_terreas = df_medias.loc[df_medias["identificador_suspense"] == 1, "Favorites"].iloc[0]
    media_casas_nao_terreas = df_medias.loc[df_medias["identificador_suspense"] == 0, "Favorites"].iloc[0]

    medias_diferenca[i] = (media_casas_terreas - media_casas_nao_terreas)

  limite_inferior = np.percentile(medias_diferenca, 2.5)
  limite_superior = np.percentile(medias_diferenca, 97.5)

  return (limite_inferior, limite_superior)
```

```python
aplicar_teste_permutacao_animes_suspense()
```

A hipótese nula será rejeitada -> Pessoas tem uma tendência para suspense
