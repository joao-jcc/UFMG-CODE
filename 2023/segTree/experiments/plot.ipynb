```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the data from the file
file_path = "results/segTree.csv"
dataST = pd.read_csv(file_path)
dataST = dataST.sort_values(by='size').reset_index()
dataST
```

```python
# Read the data from the file
file_path = "results/simpleVector.csv"
dataSV = pd.read_csv(file_path)
dataSV = dataSV.sort_values(by='size').reset_index()
dataSV
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Plotting time_query for dataST and dataSV
sns.lineplot(dataST, x='size', y='queryTime', label='SegTree')
sns.lineplot(dataSV, x='size', y='queryTime', label='SimpleVector')
plt.xlabel('Size')
plt.ylabel('Time (seconds)')
plt.title('Comparison of Query Time for SegTree and SimpleVector (1K)')
plt.legend()

# Save the plot
plt.savefig('queryTime1k.png', bbox_inches='tight')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Plotting time_query for dataST and dataSV
sns.lineplot(dataST, x='size', y='updateTime', label='SegTree')
sns.lineplot(dataSV, x='size', y='updateTime', label='SimpleVector')
plt.xlabel('Size')
plt.ylabel('Time (seconds)')
plt.title('Comparison of Update Time for SegTree and SimpleVector (1K)')
plt.legend()
# Save the plot
plt.savefig('updateTime1k.png', bbox_inches='tight')
plt.show()
```

```python
# Read data for each sorting algorithm
sorting_algorithms = ['bubble_sort', 'selection_sort', 'insertion_sort', 'quick_sort', 'merge_sort', 'heap_sort', 'count_sort']

# Create an empty DataFrame to store all data
all_data = pd.DataFrame()

# Read and concatenate data for each sorting algorithm
for algorithm in sorting_algorithms:
    file_path = f"results/{algorithm}.txt"
    data = pd.read_csv(file_path, delim_whitespace=True, names=['Size', 'Sorting Time', 'Unity'])
    data['Algorithm'] = algorithm
    all_data = pd.concat([all_data, data])

# Create a scatter plot using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=all_data, x='Size', y='Sorting Time', hue='Algorithm', style='Algorithm', markers=True, legend='full')

# Set labels and title
plt.xlabel('Graph Size')
plt.ylabel('Sorting Time (sec)')
plt.title('Sorting Time vs. Graph Size')

# Show the legend
plt.legend(title='Algorithm', loc='upper left', bbox_to_anchor=(0, 1), borderaxespad=0.)


# Save the plot
plt.savefig('sorting_time_vs_graph_size.png', bbox_inches='tight')
# Show the plot
plt.show()
```
