```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Read the data from the file
file_path = "results/merge_sort.txt"
data = pd.read_csv(file_path, delim_whitespace=True, names=['Size', 'Sorting Time', 'Unity'])
data = data.sort_values(by='Size').reset_index()
data
```

```python

# Create a scatter plot using seaborn
plt.scatter(data['Size'], data['Sorting Time'], s=2)

# Set labels and title
plt.xlabel('Number')
plt.ylabel('Sorting Time (sec)')
plt.title('Sorting Time vs. Number')

# Show the plot
plt.show()
```

```python
time = data['Sorting Time']
time
```

```python
time = time.to_list()
```

```python
derivadas = []
for i in range(1, len(time)-1):
    derivadas.append((time[i] - time[i-1])/100)
```

```python
plt.plot(derivadas)
```
