```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Read the data from the file
file_path = "results/heap_sort.txt"
data = pd.read_csv(file_path, delim_whitespace=True, names=['Size', 'Sorting Time', 'Unity'])
data.sort_values(by='Size').reset_index()
```

```python

# Create a scatter plot using seaborn
plt.plot(data['Size'], data['Sorting Time'])

# Set labels and title
plt.xlabel('Number')
plt.ylabel('Sorting Time (sec)')
plt.title('Sorting Time vs. Number')

# Show the plot
plt.show()
```
